## Example
Looking for an example? Here's what our course assistant put together for the **Ann Arbor, MI, USA** area using **sports and athletics** as the topic. [Example Solution File](./readonly/Assignment4_example.pdf)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker

dfMain = pd.read_csv('datasets_557629_1428172_covid_19_india.csv')
dfTest = pd.read_csv('datasets_557629_1428172_StatewiseTestingDetails.csv')

#Cleaning Main Data about Confirmed-Deaths-Cured cases

dfMain['Date'] = pd.to_datetime(dfMain['Date'],format='%d/%m/%y')
dfMain.drop(columns=['Sno','ConfirmedIndianNational','ConfirmedForeignNational','Time'],inplace=True)
dfMain.rename(columns={'State/UnionTerritory':'State'},inplace=True)
dfMain['MonthDay'] = dfMain['Date'].dt.strftime('%d-%m')
dfMain['Month'] = dfMain['Date'].dt.month

df1 = dfMain.copy()

#Taking data of Maharashtra & Andhra Pradesh states from unlock 1.0 onwards i.e., 1st June 2020

dfMain = dfMain[(dfMain['Month'] >= 6) & ((dfMain['State'] == 'Andhra Pradesh') | (dfMain['State'] == 'Maharashtra'))]
dfMain.set_index(['Date','State'],inplace=True)
dfMain.drop(columns=['MonthDay','Month'],inplace=True)

#Cleaning Test Data about Test Samples-Negative-Positive

dfTest['Date'] = pd.to_datetime(dfTest['Date'])
dfTest['MonthDay'] = dfTest['Date'].dt.strftime('%d-%m')
dfTest['Month'] = dfTest['Date'].dt.month

df2 = dfTest.copy()

dfTest = dfTest[(dfTest['Month'] >= 6) & ((dfTest['State'] == 'Andhra Pradesh') | (dfTest['State'] == 'Maharashtra'))]
dfTest.set_index(['Date','State'],inplace=True)
dfTest.drop(columns=['MonthDay','Month'],inplace=True)

#Preparing final data frame containing both the data

dfFinal = pd.merge(dfMain,dfTest,left_index=True,right_index=True,how='left')
dfFinal = dfFinal.reset_index()
dfFinal['TotalSamples'] = dfFinal['TotalSamples'].fillna(method='ffill')

countryConfirmedMean = df1[(df1['Month'] >= 6)].groupby(['Date'])['Confirmed'].mean()
countrySamplesMean = df2[(df2['Month'] >= 6)].groupby(['Date'])['TotalSamples'].mean()

#Picturization starts

dateRange= dfMain.reset_index()['Date'].unique()
fig = plt.figure(figsize=(10, 8), dpi=100, facecolor='w', edgecolor='k')
fig.add_subplot(2,1,1)

#Cases Details picturization

confirmedAP = dfFinal[dfFinal['State'] == 'Andhra Pradesh']['Confirmed']
confirmedMH = dfFinal[dfFinal['State'] == 'Maharashtra']['Confirmed']

plt.plot(dateRange,confirmedAP,color='lightcoral', linewidth=2) 
plt.plot(dateRange,confirmedMH,color='skyblue', linewidth=2) 
plt.plot(dateRange,countryConfirmedMean,color='green',linestyle='-',linewidth=2)

#Changing x axis to Month format

myFmt = mdates.DateFormatter('%d-%b')
plt.gca().xaxis.set_major_formatter(myFmt)

# adding legends, titles, labels & fill
ax1 = plt.gca()
ax1.set_xlabel('Date')
ax1.set_ylabel('Confirmed cases')
ax1.set_title('Progression of confirmed cases in AP & MH, India')
plt.legend(['Confirmed Cases in Andhra Pradesh', 'Confirmed Cases in Maharashtra','Mean of Confirmed Cases in India'],frameon=False)
plt.fill_between(dateRange,confirmedAP, confirmedMH,facecolor='grey', alpha=0.25)

#Testing Details picturization

fig.add_subplot(2,1,2)

testSamplesAP = dfFinal[dfFinal['State'] == 'Andhra Pradesh']['TotalSamples']
testSamplesMH = dfFinal[dfFinal['State'] == 'Maharashtra']['TotalSamples']

plt.plot(dateRange,testSamplesAP,color='lightcoral', linewidth=2,linestyle='--') 
plt.plot(dateRange,testSamplesMH,color='skyblue', linewidth=2,linestyle='--') 
plt.plot(dateRange,countrySamplesMean,color='green',linestyle='-',linewidth=2)

# adding legends, titles, labels & fill
ax2 = plt.gca()
ax2.set_xlabel('Date')
ax2.set_ylabel('Test Samples')
ax2.set_title('Progression of Test Samples in AP & MH, India')
plt.legend(['Test Samples in Andhra Pradesh', 'Test Samples in Maharashtra','Mean of Test Samples in India'],frameon=False)
plt.fill_between(dateRange,testSamplesAP, testSamplesMH,facecolor='grey', alpha=0.25)

#Changing x axis to Month format

myFmt = mdates.DateFormatter('%d-%b')
plt.gca().xaxis.set_major_formatter(myFmt)

#Changing y axis format

ax1.get_yaxis().set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
ax2.get_yaxis().set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

#Adjusting spaces between plots

plt.subplots_adjust(top = 0.99, bottom=0.01)

```
